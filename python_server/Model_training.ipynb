```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, mean_absolute_error
from xgboost import XGBClassifier, XGBRegressor

# Start from your df (with token_id)
df_base = df.copy()

# -------------------------------------------------
# 0. Sort by token + time for meaningful lags
# -------------------------------------------------
df_base["timestamp"] = pd.to_datetime(df_base["timestamp"])
df_base = df_base.sort_values(["token_id", "timestamp"]).reset_index(drop=True)

# -------------------------------------------------
# 1. Feature Engineering: lagged / rolling per token
# -------------------------------------------------
def add_group_features(g):
    g = g.copy()
    g["rsi_1"] = g["rsi"].shift(1)
    g["rsi_3"] = g["rsi"].rolling(3).mean()
    g["rsi_5"] = g["rsi"].rolling(5).mean()

    g["atr_3"] = g["atr"].rolling(3).mean()
    g["atr_5"] = g["atr"].rolling(5).mean()

    g["volume_avg_3"] = g["volume_avg"].rolling(3).mean()
    g["volume_avg_5"] = g["volume_avg"].rolling(5).mean()
    g["obv_3"] = g["obv"].rolling(3).mean()

    return g

df_fe = df_base.groupby("token_id", group_keys=False).apply(add_group_features)

# Fill NaNs from shifts/rolling
df_fe.fillna(method="bfill", inplace=True)
df_fe.fillna(method="ffill", inplace=True)

# -------------------------------------------------
# 2. Encode categorical targets/features
# -------------------------------------------------
le_supertrend = LabelEncoder()
df_fe["supertrend_trend_enc"] = le_supertrend.fit_transform(df_fe["supertrend_trend"])

le_dir = LabelEncoder()
df_fe["extreme_type_enc"] = le_dir.fit_transform(df_fe["extreme_type"])

# -------------------------------------------------
# 3. Feature columns (no token_id, no targets)
# -------------------------------------------------
feature_cols = [
    "rsi", "ema10", "ema20", "ema50", "ema100", "ema200", "ema_cross",
    "macd_line", "macd_signal", "macd_hist", "vwap", "adx", "plus_di",
    "minus_di", "stoch_k", "stoch_d", "boll_upper", "boll_middle",
    "boll_lower", "boll_percent", "atr", "obv", "supertrend_value",
    "cci", "roc", "momentum3", "volume_sum", "volume_avg",
    "range_pct", "open_rel", "high_rel", "low_rel", "close_rel",
    "hour", "minute", "weekday",
    # engineered features
    "rsi_1", "rsi_3", "rsi_5",
    "atr_3", "atr_5",
    "volume_avg_3", "volume_avg_5", "obv_3",
    "supertrend_trend_enc",
]

X_all = df_fe[feature_cols]
y_dir_all = df_fe["extreme_type_enc"]
y_pct_all = df_fe["pct_change"]

# log-transform % change
y_pct_log_all = np.log1p(np.abs(y_pct_all)) * np.sign(y_pct_all)

# -------------------------------------------------
# 4. Token-based train/test split (CRITICAL)
# -------------------------------------------------
tokens = df_fe["token_id"].unique()
train_tokens, test_tokens = train_test_split(
    tokens, test_size=0.2, random_state=42
)

train_mask = df_fe["token_id"].isin(train_tokens)
test_mask  = df_fe["token_id"].isin(test_tokens)

X_train_dir = X_all[train_mask]
X_test_dir  = X_all[test_mask]
y_train_dir = y_dir_all[train_mask]
y_test_dir  = y_dir_all[test_mask]

# (optional) shuffle within train only
X_train_dir, y_train_dir = shuffle(
    X_train_dir, y_train_dir, random_state=42
)

# -------------------------------------------------
# 5. Train Direction Classifier
# -------------------------------------------------
clf_dir = XGBClassifier(
    n_estimators=400,           # you can increase later
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="mlogloss",
    tree_method="hist",
    random_state=42,
)
clf_dir.fit(X_train_dir, y_train_dir)
y_pred_dir = clf_dir.predict(X_test_dir)

print("=== Direction Classification (token-based split) ===")
print(classification_report(y_test_dir, y_pred_dir, target_names=le_dir.classes_))

# -------------------------------------------------
# 6. Train % Change Regressor (Hybrid, only up/down, train tokens only)
# -------------------------------------------------
stable_enc = le_dir.transform(["stable"])[0]
mask_up_down_train = train_mask & (df_fe["extreme_type_enc"] != stable_enc)

X_up_down_train = X_all[mask_up_down_train].copy()
y_up_down_pct_log_train = y_pct_log_all[mask_up_down_train]

# use true direction encoding as feature in training
X_up_down_train["pred_dir_feature"] = df_fe.loc[mask_up_down_train, "extreme_type_enc"]

X_train_pct = X_up_down_train
y_train_pct = y_up_down_pct_log_train

reg_pct = XGBRegressor(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method="hist",
    random_state=42,
)
reg_pct.fit(X_train_pct, y_train_pct)

# -------------------------------------------------
# 7. Hybrid prediction on TEST TOKENS only
# -------------------------------------------------
y_pred_pct_hybrid = np.zeros(len(X_test_dir))
stable_enc = le_dir.transform(["stable"])[0]

for idx, row_idx in enumerate(X_test_dir.index):
    direction = y_pred_dir[idx]

    if direction != stable_enc:
        X_row = X_test_dir.loc[[row_idx]].copy()
        X_row["pred_dir_feature"] = direction
        y_pred_pct_hybrid[idx] = reg_pct.predict(X_row)[0]
    else:
        y_pred_pct_hybrid[idx] = 0.0

# reverse log-transform
y_pred_pct_hybrid_inv = np.sign(y_pred_pct_hybrid) * (np.expm1(np.abs(y_pred_pct_hybrid)))
y_test_pct_actual = y_pct_all.loc[X_test_dir.index]

mae_pct_hybrid = mean_absolute_error(y_test_pct_actual, y_pred_pct_hybrid_inv)
print("=== % Change Regression (Hybrid, token-based split) ===")
print("MAE % Change:", mae_pct_hybrid)
```
